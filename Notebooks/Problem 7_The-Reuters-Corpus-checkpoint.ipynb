```python
# Importing the required libraries for doing sentiment analysis and training a predictive model i.e. Naive bayes
import os
import pandas as pd
import numpy as np
import scipy
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Download NLTK data
nltk.download('vader_lexicon')
```

```python
# Loading the training data
train_path = 'C:/Users/krish/Downloads/STA380-master/data/ReutersC50/C50train'
authors_train = os.listdir(train_path)
documents_train = []
labels_train = []

for author in authors_train:
    author_train_path = os.path.join(train_path, author)
    for file in os.listdir(author_train_path):
        with open(os.path.join(author_train_path, file), 'r', encoding='latin-1') as f:
            content = f.read()
            documents_train.append(content)
            labels_train.append(author)
```

```python
# Create lists to store document content and author labels for training data
train_path = 'C:/Users/krish/Downloads/STA380-master/data/ReutersC50/C50train'
documents_train = []
authors_train = []

# Load data
for author in os.listdir(train_path):
    author_train_path = os.path.join(train_path, author)
    if os.path.isdir(author_train_path):  # Check if the item is a directory
        for file in os.listdir(author_train_path):
            with open(os.path.join(author_train_path, file), 'r', encoding='latin-1') as f:
                content = f.read()
                documents_train.append(content)
                authors_train.append(author)

# Create a DataFrame from the data
data_train = pd.DataFrame({'Author': authors_train, 'Document': documents_train})
```

```python
#Loading the test data
test_path = 'C:/Users/krish/Downloads/STA380-master/data/ReutersC50/C50test'
authors_test = os.listdir(test_path)
documents_test = []
labels_test = []

for author in authors_test:
    author_test_path = os.path.join(test_path, author)
    for file in os.listdir(author_test_path):
        with open(os.path.join(author_test_path, file), 'r', encoding='latin-1') as f:
            content = f.read()
            documents_test.append(content)
            labels_test.append(author)  
```

Let's perform sentiment analysis on this dataset.

```python
# Perform sentiment analysis using NLTK's SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
sentiments_train = [sia.polarity_scores(doc)['compound'] for doc in documents_train]
sentiments_test = [sia.polarity_scores(doc)['compound'] for doc in documents_test]

# Creating a dataframe of Document - Author - Sentiment_Scores
data_train = pd.DataFrame({'Document': documents_train, 'Author': labels_train, 'Sentiment': sentiments_train})
data_test = pd.DataFrame({'Document': documents_test, 'Author': labels_test, 'Sentiment': sentiments_test})
```

### Training Multinomial Naive Bayes' classifer to predict sentiment of a document using `Content` and `Author`

```python
# Convert text data to TF-IDF features
vectorizer = TfidfVectorizer(max_features=1500)  
X_train_content = vectorizer.fit_transform(data_train['Document'])
X_train_author = vectorizer.fit_transform(data_train['Author'])
X_train = scipy.sparse.hstack((X_train_content, X_train_author))

X_test_content = vectorizer.fit_transform(data_test['Document'])
X_test_author = vectorizer.fit_transform(data_test['Author'])
X_test = scipy.sparse.hstack((X_test_content, X_test_author))

y_train = ['positive' if sentiment > 0.5 else 'negative' if sentiment < -0.25 else 'neutral' for sentiment in data_train['Sentiment']]
y_test = ['positive' if sentiment > 0.5 else 'negative' if sentiment < -0.25 else 'neutral' for sentiment in data_test['Sentiment']]
```

```python
# Plot the number of positive, negative, and neutral documents for both training and test data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # Subplot for training data
train_sentiment_counts = pd.Series(y_train).value_counts()
train_sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Training Data: Number of Positive, Negative, and Neutral Documents')
plt.xlabel('Sentiment')
plt.ylabel('Number of Documents')
for index, value in enumerate(train_sentiment_counts):
    plt.text(index, value + 1, str(value), ha='center', va='bottom')
    
plt.subplot(1, 2, 2)  # Subplot for test data
test_sentiment_counts = pd.Series(y_test).value_counts()
test_sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Test Data: Number of Positive, Negative, and Neutral Documents')
plt.xlabel('Sentiment')
plt.ylabel('Number of Documents')
for index, value in enumerate(test_sentiment_counts):
    plt.text(index, value + 1, str(value), ha='center', va='bottom')
    
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Train a Naive Bayes classifier
model_nb = MultinomialNB()
model_nb.fit(X_train, y_train)

# Predict sentiment on the test set
y_pred_train = model_nb.predict(X_train)
y_pred_test = model_nb.predict(X_test)

# Evaluate the model
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

conf_matrix = confusion_matrix(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Plot confusion matrix using seaborn and matplotlib
Sentiment_types = ['Negative', 'Neutral', 'Positive']
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=Sentiment_types, yticklabels=Sentiment_types)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

### Training Multinomial Naive Bayes' classifer to predict sentiment of a document using `Content`

```python
# Convert text data to TF-IDF features
vectorizer = TfidfVectorizer(max_features=1000)  
X_train = vectorizer.fit_transform(data_train['Document'])
X_test = vectorizer.fit_transform(data_test['Document'])

y_train = ['positive' if sentiment > 0.5 else 'negative' if sentiment < -0.25 else 'neutral' for sentiment in data_train['Sentiment']]
y_test = ['positive' if sentiment > 0.5 else 'negative' if sentiment < -0.25 else 'neutral' for sentiment in data_test['Sentiment']]
```

```python
# Train a Naive Bayes classifier
model_nb = MultinomialNB()
model_nb.fit(X_train, y_train)

# Predict sentiment on the test set
y_pred_train = model_nb.predict(X_train)
y_pred_test = model_nb.predict(X_test)

# Evaluate the model
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)
conf_matrix = confusion_matrix(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Plot confusion matrix using seaborn and matplotlib
Sentiment_types = ['Negative', 'Neutral', 'Positive']
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=Sentiment_types, yticklabels=Sentiment_types)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Get the log probabilities of features for each class
log_probabilities = model_nb.feature_log_prob_

# Convert the log probabilities to probabilities
probabilities = np.exp(log_probabilities)

# Get the feature names
feature_names = vectorizer.get_feature_names_out()

# Create a DataFrame for the probabilities
probabilities_df = pd.DataFrame(probabilities, columns=feature_names, index=model_nb.classes_)

# Get the top features for each class
num_top_features = 10  # You can adjust the number of top features to display
top_features = {}
for label in model_nb.classes_:
    top_features[label] = probabilities_df.loc[label].sort_values(ascending=False)[:num_top_features]

# Print the top features for each class
for label, features in top_features.items():
    print(f"Top features for {label}:")
    print(features)
    print()
```

#### Conclusion

After careful analysis of the results, it becomes evident that incorporating author names as additional features to train the model has a detrimental effect on the overall accuracy. This decline in accuracy can be attributed to the diverse content associated with each author, leading the model to struggle in effectively distinguishing between the positive and negative sentiment. The inclusion of author information introduces a confounding factor that obscures the model's ability to discern the true sentiment of the documents.

A closer look at the confusion matrices for both models underscores this point. The matrices highlight a prevalent trend of predicting positive sentiment across a significant portion of documents. This behavior closely mirrors the outcomes of a baseline classifier, where the positive sentiment is seemingly overrepresented. Such a pattern is likely a consequence of the model's difficulties in disentangling the complex interplay between authorship, content, and sentiment.

This inference is substantiated by the analysis of feature importance. Notably, the top features for all three sentiment classes are remarkably consistent, hinting at a lack of discriminative power among the features. This phenomenon reinforces the hypothesis that the introduction of author names has introduced noise into the model's learning process, leading it to place undue importance on these extraneous cues.
